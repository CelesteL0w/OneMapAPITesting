Run API for primary schools after deleting columns

```python
file = pd.read_csv("./Indiv dataset/Primary school V3.csv")
s = file["postal_code"]
```

```python
print("One map api")

for i in range(len(s)):
#     i += 123
    print(i)
    query_string='https://developers.onemap.sg/commonapi/search?searchVal='+str(s[i])+'&returnGeom=Y&getAddrDetails=Y&pageNum=1'
    resp = requests.get(query_string)
    
    #Convert JSON into Python Object
    print("Converting to python obj")
    data=json.loads(resp.content)
    
    # Add values into the lang and lon list
    latitude_list.append(data['results'][0]['LATITUDE'])
    longitude_list.append(data['results'][0]["LONGITUDE"])

print('Done.')
```
